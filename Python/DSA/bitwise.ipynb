Find the given no is even or odd using bitwise operator

```python
def even_odd(n):
    if (n & 1):
        return "odd"
    else:
        return "even"
even_odd(11)
```

Given an array of n element, every element of array is repeating except one find that element

```python
def non_repeating(arr):
    a = 0
    for i in arr:
        a^=i
    return a
non_repeating([2,1,3,2,1,3,8,8,9])
```

Hamming distance

```python
def hamming_dist(n,m):
    flipbits = n ^ m
    c = 0
    #print(flipbits)
    while flipbits:
        c += flipbits & 1
        flipbits >>= 1
    return c
hamming_dist(10,20)
        
        
```

Return true if it is in power 2 else false

```python
def power2(n):
    if n <=  0:
        return False
    while n != 1:
        if n % 2 != 0:
            return False
        n = n // 2
    return True
power2(32)
```

```python
def power_two(n):
    if n <= 0:
        return False
    return n & (n-1) == 0
power_two(64)
```
