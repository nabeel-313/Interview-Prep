Given an array of size N, given a number K, return the  max sum of subarray of size K
N = 5, K=3, array= [1,3,5,2,4]
ans = 11

```python

array= [1,2,3,4,3,2,1,0]
def max_sum_subarray(array):
    n = 0
    K= 4
    N = len(array)
    sum = []
    while n < N-K:
        sub_array = array[n:n+K]
        #print(sub_array)
        su = 0
        for i in sub_array:
            su = su +i
        sum.append(su)
        n = n+1
    return max(sum)
max_sum_subarray(array)
```

Given an m x n integer matrix matrix, if an element is 0, set its entire row and column to 0's.

```python
mtrx = [[1,1,1],[1,0,1],[1,1,2]]
def mtrx_zero(matrix):
    print(matrix)
    rows = len(matrix)
    cols = len(matrix[0])
    zero_rows = set()
    zero_cols = set()

    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] == 0:
                zero_rows.add(i)
                zero_cols.add(j)
    #Set the identified rows to zero
    for row in zero_rows:
        for j in range(cols):
            matrix[row][j] = 0

    #Set the identified columns to zero
    for col in zero_cols:
        for i in range(rows):
            matrix[i][col] = 0
    #print(matrix)
    for mat in matrix:
        print(mat)

mtrx_zero(mtrx)
```

Given an array of strings strs, group the anagrams together. You can return the answer in any order.

```python
from collections import defaultdict

def groupAnagrams(strs):
    anagrams = defaultdict(list)  # Create a dictionary to hold lists of anagrams

    for s in strs:
        # Sort the string and use it as a key
        sorted_str = ''.join(sorted(s))
        anagrams[sorted_str].append(s)

    # Return the grouped anagrams as a list of lists
    return list(anagrams.values())

# Example usage
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
grouped_anagrams = groupAnagrams(strs)
print(grouped_anagrams)
```

You are given an array prices where prices[i] is the price of a given stock on the ith day.

You want to maximize your profit by choosing a single day to buy one stock and choosing a different day in the future to sell that stock.

Return the maximum profit you can achieve from this transaction. If you cannot achieve any profit, return 0.

```python
prices = [7,6,4,3,1]
def max_profit(price):
    profit = []
    for i in range(len(price)): #buy
        for j in range(i+1, len(price)-1): #sell
            prof = price[j] - price[i]
            profit.append(prof)
    if max(profit) < 0:
        return 0
    else:
        return max(profit)
max_profit(prices)
```

You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store

height = [1,8,6,2,5,4,8,3,7]
Output: 49

```python
height = [1,8,6,2,5,4,8,3,7]
def max_water(height):
    area = []
    for i in range(len(height)):
        for j in range(i+1,len(height)):
            h = min(height[i], height[j])
            w = j-i
            area.append(h*w)
    return max(area)
max_water(height)


```

Given an array nums with n objects colored red, white, or blue, sort them in-place so that objects of the same color are adjacent, with the colors in the order red, white, and blue.

We will use the integers 0, 1, and 2 to represent the color red, white, and blue, respectively.

nums = [2,0,2,1,1,0]
Output: [0,0,1,1,2,2]

```python
n = [2,0,2,1,1,0]
def sortcolor(nums):
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] > nums[j]:
                nums[i], nums[j] = nums[j], nums[i]
    return nums
sortcolor(n)
```

Given an array of positive integers nums and a positive integer target, return the minimal length of a
subarray
whose sum is greater than or equal to target. If there is no such subarray, return 0 instead.

 target = 7, nums = [2,3,1,2,4,3]
Output: 2

```python
nums = [2,3,1,2,4,3]
target = 8

def min_subarray(num,tar):
    min_len = len(num) + 1
    for i in range(len(num)):
        cur_sum = num[i]
        for j in range(i+1,len(num)):
            cur_sum = cur_sum + num[j]

            if cur_sum >= tar and (j - i + 1) < min_len:
                min_len = j-i+1
    return min_len
min_subarray(nums, target)



```

You are given two integer arrays nums1 and nums2, sorted in non-decreasing order, and two integers m and n, representing the number of elements in nums1 and nums2 respectively.

Merge nums1 and nums2 into a single array sorted in non-decreasing order.

Input: nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3
Output: [1,2,2,3,5,6]

```python
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3

def merge(nums1, m, nums2, n):
    # Initialize pointers for nums1 and nums2
    i, j = 0, 0
    # Make a copy of the first m elements of nums1
    nums1_copy = nums1[:m]

    k = 0  # Pointer for the merged array

    # Merge nums1_copy and nums2 into nums1
    while i < m and j < n:
        if nums1_copy[i] < nums2[j]:
            nums1[k] = nums1_copy[i]
            i += 1
        else:
            nums1[k] = nums2[j]
            j += 1
        k += 1

    # If there are remaining elements in nums1_copy
    while i < m:
        nums1[k] = nums1_copy[i]
        i += 1
        k += 1

    # If there are remaining elements in nums2
    while j < n:
        nums1[k] = nums2[j]
        j += 1
        k += 1

# Example usage:

merge(nums1, m, nums2, n)
print(nums1)
```

Given an m x n matrix, return all elements of the matrix in spiral order.

Input: matrix = [[1,2,3],[4,5,6],[7,8,9]]

Output: [1,2,3,6,9,8,7,4,5]

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
def spiralOrder(matrix):
    res = []
    if len(matrix) == 0:
        return res
    row_begin = 0
    col_begin = 0
    row_end = len(matrix)-1
    col_end = len(matrix[0])-1
    while (row_begin <= row_end and col_begin <= col_end):
        for i in range(col_begin,col_end+1):
            res.append(matrix[row_begin][i])
        row_begin += 1
        for i in range(row_begin,row_end+1):
            res.append(matrix[i][col_end])
        col_end -= 1
        if (row_begin <= row_end):
            for i in range(col_end,col_begin-1,-1):
                res.append(matrix[row_end][i])
            row_end -= 1
        if (col_begin <= col_end):
            for i in range(row_end,row_begin-1,-1):
                res.append(matrix[i][col_begin])
            col_begin += 1
    return res
spiralOrder(matrix)
```

```python
# Sudoko Validator
board = [["5","3",".",".","7",".",".",".","."]
,["6",".",".","1","9","5",".",".","."]
,[".","9","8",".",".",".",".","6","."]
,["8",".",".",".","6",".",".",".","3"]
,["4",".",".","8",".","3",".",".","1"]
,["7",".",".",".","2",".",".",".","6"]
,[".","6",".",".",".",".","2","8","."]
,[".",".",".","4","1","9",".",".","5"]
,[".",".",".",".","8",".",".","7","9"]]
def isValidSudoku(board) -> bool:
    result = []
    for i in range(9):
        for j in range(9):
            ele = board[i][j]
            if ele != '.':
                result += [(i,ele) , (ele,j), (i // 3, j // 3, ele)]
    return len(result) == len(set(result))
isValidSudoku(board)
```
