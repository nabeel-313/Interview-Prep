```python
## Find the min value of unsorted array
min_arr = [4,9,3,7,1,8,2]
def mn_ele(arr):
    min_ele = arr[0]
    for n in arr:
        if min_ele > n:
            min_ele = n
    return min_ele
mn_ele(min_arr)
```

```python
## find the value which is equal to its index 1 based index
arrr = [7,9,6,4,1,8,2]
def val_ind(ar):
    for i in range(len(ar)):
        if i+1 == ar[i]:
            return ar[i]
val_ind(arrr)
        
```
