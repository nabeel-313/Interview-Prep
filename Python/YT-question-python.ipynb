```python
# remove suffix -->> ing, ed
def remove_suff(inp):
    res_list = []
    n_list = inp.split(" ")
    for w in n_list:
        temp = w
        if w[-3:] == 'ing':
            temp = w[:-3]
        elif w[-2:] == 'ed':
            temp = w[:-2]
        res_list.append(temp)
    res = " ".join(res_list)
    return res
n = ' A boy is playing or played'
remove_suff(n)
```

```python
# total no of anagram
def total_anagram(ana):
    cnt = {}
    for i in ana:
        sorted_i = ''.join(sorted(i))
        if sorted_i in cnt:
            cnt[sorted_i] += 1
        else:
            cnt[sorted_i] = 1
    print(cnt)
    return len(cnt)
ana = ['ate','inch', 'tea', 'nich', 'cat','act']
print("total ana is ",total_anagram(ana))

from collections import defaultdict

def groupAnagrams(strs):
    anagrams = defaultdict(list)  # Create a dictionary to hold lists of anagrams

    for s in strs:
        # Sort the string and use it as a key
        sorted_str = ''.join(sorted(s))
        anagrams[sorted_str].append(s)

    # Return the grouped anagrams as a list of lists
    return list(anagrams.values())

# Example usage
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
grouped_anagrams = groupAnagrams(strs)
print(grouped_anagrams)
```

```python
#pattern matching return true of there 6-12 y after x
def patter(inp):
    for i in range(len(inp)-1):
        if inp[i] == 'x':
            cnt = 0
            for j in range(i+1,len(inp)):
                if inp[j] == 'y':
                    cnt = cnt + 1
                else :
                    break
            if 6 <= cnt <= 12:
                return True
    return False
patter('xyyy')
```

```python
#paranthesischeck
def para(exp):
    stack = []
    for p in exp:
        if p in ['(','[','{']:
            stack.append(p)
        else:
            if not stack:
                return False
            curr = stack.pop()
            if curr == '(' :
                if p != ')':
                    return False
            if curr == '[' :
                if p != ']':
                    return False
            if curr == '{' :
                if p != '}':
                    return False
    if stack:
        return False
    return True
para('[({(})]')
```

```python
arr = [3,5,8,12,15]
tar = 7
def correct_insert(arr, n):
    ind = len(arr)
    for i in range(len(arr)):
        if arr[i] > n:
            ind = i
            break
    if ind == len(arr):
        arr = arr[:ind] + [n]
    else:
        #arr = arr[:ind] + [n] + arr[ind:]
        arr.insert(ind, n)
    return arr
correct_insert(arr,tar)
```

```python
def palindrom(st):
    st = st.lower()
    i = 0
    j = len(st) -1
    while i <=j :
        if st[i] == st[j]:
            i +=1
            j -=1
        else:
            return False
    return True
palindrom('Nitmn')
```

```python
# find pair of 3 whhose multiple is 180
def pair_3(lst, tar):
    triplet = []
    for i in range(len(lst)):
        for j in range(i+1, len(lst)):
            third = tar // (lst[i] * lst[j])
            if tar % (lst[i] * lst[j]) == 0 and third in lst[j+1:]:
                triplet.append((lst[i], lst[j], third))

    return triplet

pair_3([15,6,2,3,10,8,7,9],180)
```

```python
import pandas as pd
from itertools import combinations

# Create a DataFrame of teams
data = {
    'team_id': [1, 2, 3, 4, 5],
    'team_name': ['Team A', 'Team B', 'Team C', 'Team D', 'Team E']
}

df = pd.DataFrame(data)

# Generate all unique pairs of teams
team_pairs = list(combinations(df['team_id'], 2))

# Create a new DataFrame with the pairs
matches = pd.DataFrame(team_pairs, columns=['team1', 'team2'])

print(matches)
```

```python
# must contain one lower case, uppercase, special char, len = 8
def pswd_check(psw):
    lower = False
    upper = False
    dgt = False
    special_char = False

    sp_char = '!@#$%&<>?'
    for i in range(len(psw)):
        if psw[i].islower():
            lower = True
        if psw[i].isupper():
            upper = True
        if psw[i].isdigit():
            dgt = True
        if psw[i] in sp_char:
            special_char = True


    if (lower and upper and dgt and special_char and len(psw) > 7 ):
        return True
    else:
        return False
pswd_check('Nabeel123&')
```
