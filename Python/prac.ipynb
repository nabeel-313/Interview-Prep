```python
def rev_string(st):
    st = st.split()
    rev_st = []
    for s in st:
        rev_st.insert(0,s)
    return ' '.join(rev_st)

def rev_word(word):
    word = list(word)
    l, r = 0, len(word)-1
    while l < r:
        word[l], word[r] = word[r], word[l] 
        l += 1
        r -= 1
    return ''.join(word)

def is_palidrom(wrd):
    l,r = 0, len(wrd)-1
    while l < r:
        if wrd[l] == wrd[r]:
            l += 1
            r -= 1
        else:
            return False
    return True

print(rev_string("My name is nabeel"))
print(rev_word("Nabeel"))
print(is_palidrom("MadaM"))
```

```python
from collections import defaultdict

def anagram_together(lst):
    ana = defaultdict(list)
    for w in lst:
        w_sorted = ''.join(sorted(w))
        ana[w_sorted].append(w)

    return ana.values()
anagram_together(['ate','inch', 'tea', 'nich', 'cat','act'])
```

```python
def same_len_wrd(sent):
    out = {}
    word = sent.split()
    for w in word:
        ln = len(w)
        if ln not in out:
            out[ln] = []
        out[ln].append(w)
    return out
same_len_wrd("My name is nabeel ahmed and I am an Machine Learning Engineer")
```

```python
nums = [-1,0,3,5,9,12]; target = 9
def bin_search(num, tar):
    l = 0
    r = len(num)-1
    while l <= r:
        mid = (l+r)//2
        if num[mid] == tar:
            return mid
        elif num[mid] < tar:
            l = mid + 1
        else:
            r = mid - 1
    return -1
bin_search(nums, target)
```

```python
#shift even left side and oddright side
inp = [1,2,3,4,5,6,7,8,9]
def shift_even_odd(arr):
    left = 0
    right = len(arr)-1
    while left < right:
        if arr[left]%2 ==0:
            left = left + 1
        elif arr[right]%2 != 0:
            right = right -1
        else:
            arr[left], arr[right] = arr[right], arr[left]
    return arr
print(shift_even_odd(inp))

def shift_even_odd1(arr):
    even = [x for x in arr if x % 2 == 0]
    odd = [x for x in arr if x % 2 != 0]
    return even + odd

inp = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(shift_even_odd1(inp))
```

```python
import time

def time_exce(funct):
    def inner(*args, **kwargs):
        start_time = time.time()
        result = funct(*args, **kwargs)
        end_time = time.time()
        proc_time = end_time - start_time
        print(f"Execution time: {proc_time:.5f} seconds")
        return result
    return inner
```

```python
@time_exce
def first_non_repeat_char(s):
    for i in range(len(s)):
        if s[i] not in s[i+1:]:
            return s[i]
first_non_repeat_char('swiss')
```

```python
def fib_gen(n):
    a, b = 0, 1  # First two Fibonacci numbers
    for _ in range(n):
        yield a  # Yield current number
        a, b = b, a + b  # Update to next Fibonacci numbers

# Example usage:
for num in fib_gen(10):  # Generate first 10 Fibonacci numbers
    print(num)
```

```python
def reverse_alpha(st):
    st = list(st)
    l, r = 0, len(st) - 1

    while l < r:
        if st[l].isalpha() and st[r].isalpha():
            st[l], st[r] = st[r], st[l]
            l += 1
            r -= 1
        elif not st[l].isalpha():
            l += 1
        elif not st[r].isalpha():
            r -= 1

    return ''.join(st)

print(reverse_alpha('n@ab&e$el#'))
```

```python
def max_asce_sum(arr):
    mx_sum = 0
    cur_sum = arr[0]
    for i in range(1,len(arr)):
        if arr[i] > arr[i-1]:
            cur_sum +=arr[i]
            mx_sum = max(mx_sum, cur_sum)
        else:
            cur_sum = arr[i]
            mx_sum = max(mx_sum, cur_sum)
    return mx_sum
max_asce_sum([10,20,30,50])
```
