## next warmday using stack monotonic

```python
arr = [74, 75, 76, 70, 71, 79, 62]

def next_warmday(arr):
    n = len(arr)
    res = [0] * n
    stk = []
    
    for i in range(n):  # Using a standard for loop
        t = arr[i]  # Extract temperature at index i
        while stk and arr[stk[-1]] < t:
            stk_i = stk.pop()
            res[stk_i] = i - stk_i
        stk.append(i)
    
    return res

print(next_warmday(arr))
```

```python
def next_greater_element(arr):
    n = len(arr)
    res = [-1] * n  
    stk = []  

    for i in range(n):
        while stk and arr[stk[-1]] < arr[i]:  # If current element is greater
            stk_i = stk.pop()
            res[stk_i] = arr[i]  # Store the next greater element
        stk.append(i)  # Push the current index

    return res

# Example input
arr = [74, 75, 76, 70, 71, 79, 62]
print(next_greater_element(arr))

def next_smaller_element(arr):
    n = len(arr)
    res = [-1] * n  
    stk = []  

    for i in range(n):
        while stk and arr[stk[-1]] > arr[i]:  # Change '>' to find the smaller element
            stk_i = stk.pop()
            res[stk_i] = arr[i]  # Store the next smaller element
        stk.append(i)  # Push the current index

    return res

# Example input
arr = [74, 75, 76, 70, 71, 79, 62]
print(next_smaller_element(arr))
```
