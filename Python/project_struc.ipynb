```python
import os
import yaml

def create_project_structure(config, base_path='.'):
    for folder, contents in config.items():
        folder_path = os.path.join(base_path, folder)
        os.makedirs(folder_path, exist_ok=True)

        if isinstance(contents, dict):
            for file_folder, file_name in contents.items():
                if isinstance(file_name, dict):
                    create_project_structure({file_folder: file_name}, base_path=folder_path)
                else:
                    file_path = os.path.join(folder_path, file_folder)
                    print("dict",file_path)
                    with open(file_path, 'w') as file:
                        pass
        elif isinstance(contents, str):
            # Assume it's a file, create an empty file
            file_path = os.path.join(base_path, contents)
            print("str",file_path)
            with open(file_path, 'w') as file:
                pass

if __name__ == "__main__":
    with open('struc.yaml', 'r') as file:
        project_config = yaml.safe_load(file)

    create_project_structure(project_config)
    print("Project structure created successfully.")
```
